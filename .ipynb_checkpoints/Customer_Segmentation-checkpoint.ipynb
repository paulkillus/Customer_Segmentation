## 1. Cohort Analysis

In customer segmentation, we divide customers in to mutually exclusive segments, which is called cohorts. After the segmentation, we can compare metrics across product lifecycle as well as customer lifecycle.

The most common approaches of dividing cohorts include:
- Time cohorts
- Behavior cohorts
- Size cohorts

For time cohorts analysis, we will use the earliest month when the customer purchased our product as the measurement of dividing cohorts.

<a class="anchor" id="1.1"></a>

### 1.1 Data acquisition and cleaning

First, import the packages and read the data from csv files.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import datetime as dt
```

```python
dtype = {"Invoice":str,
        "StockCode":str,
        "Description":str,
        "Quantity":float,
        "InvoiceDate":str,
        "Price":float,
        "Customer ID":str,
        "Country":str}
df_2009 = pd.read_csv("online_retail_II_2009.csv",dtype = dtype, parse_dates = [4])
df_2010 = pd.read_csv("online_retail_II_2010.csv",dtype = dtype, parse_dates = [4])
df = pd.concat([df_2009,df_2010])
print(df.shape)
df.head()
```

```python
df.info()
```

We noticed that some data in the columns 'Desciption' and 'Customer ID' are missing. Furthur inspectation shows that some rows have no Customer ID information, and other features are all invalid information. We assume that these are non-customer related transaction, so we decide to delete the rows without Customer ID.

```python
df[df["Description"].isnull()].head()
```

```python
df[df["Description"].isnull()]["Customer ID"].value_counts()
```

```python
df[df["Customer ID"].isnull()].head()
```

```python
df[df["Invoice"].isin(df[df["Customer ID"].isnull()]["Invoice"])]["Customer ID"].value_counts()
```

```python
# Delete all rows without a valid Customer ID
df = df[~df["Customer ID"].isnull()]
```

```python
df.info()
```

<a class="anchor" id="1.2"></a>
### 1.2 Building the cohort table

To build the cohort table, we need to create some new value as the cohort month, the time offset, etc. Then we use DataFrame.pivot_table to stack the data into a new table, which will be the basis for our analysis.

```python
def get_month(x): return dt.datetime(x.year, x.month, 1)
```

```python
df['InvoiceMonth'] = df['InvoiceDate'].apply(get_month)
```

```python
grouping = df.groupby('Customer ID')['InvoiceMonth']
```

```python
df['CohortMonth'] = grouping.transform('min')
```

```python
df.head()
```

```python
def get_date_int(df, column):
    year = df[column].dt.year
    month = df[column].dt.month
    day = df[column].dt.day
    return year, month, day
```

```python
invoice_year, invoice_month, _ = get_date_int(df, 'InvoiceMonth')
cohort_year, cohort_month, _ = get_date_int(df, 'CohortMonth')
```

```python
years_diff = invoice_year - cohort_year
months_diff = invoice_month - cohort_month
```

```python
df['CohortIndex'] = years_diff * 12 + months_diff + 1
df.head()
```

```python
cohort_counts = df.pivot_table(index='CohortMonth',columns="CohortIndex",values="Customer ID",aggfunc="count")
```

```python
cohort_counts.head()
```

<a class="anchor" id="1.3"></a>
### 1.3 Visualization of customer retention and other metrics

The heatmap is the most popular visualization of showing metrics for cohorts analysis. We calculate the metrics inlcuding the retention rate, average quantity purchased, etc. and then create the coresponding heatmaps.

```python
cohort_sizes = cohort_counts.iloc[:,0]
retention = cohort_counts.divide(cohort_sizes, axis=0)
```

```python
fig, ax = plt.subplots(figsize=(20,7))
ax.set_title("Customer Retention Rate")
sns.heatmap(data = retention,
            ax=ax,
            annot=True,
            fmt=".1%",
            vmin= 0.0,
            vmax = 0.5,
            cmap = 'BuGn',
            yticklabels=retention.index.strftime("%Y-%m"))
```

```python
grouping = df.groupby(['CohortMonth', 'CohortIndex'])
cohort_data = grouping['Quantity'].mean()
cohort_data = cohort_data.reset_index()

average_quantity = cohort_data.pivot(index='CohortMonth', 
                                     columns='CohortIndex', 
                                     values='Quantity')
```

```python
fig, ax = plt.subplots(figsize=(20,7)) 
ax.set_title("Average Quantity Purchased by Monthly Cohorts")
sns.heatmap(data = average_quantity,
            ax=ax,
            annot=True,
            fmt=".1f",
            vmin= 5,
            vmax = 30,
            cmap = 'Blues',
            yticklabels=retention.index.strftime("%Y-%m"))
```

```python
grouping = df.groupby(['CohortMonth', 'CohortIndex'])
cohort_data = grouping['Price'].mean()
cohort_data = cohort_data.reset_index()

average_price = cohort_data.pivot(index='CohortMonth', 
                                     columns='CohortIndex', 
                                     values='Price')
```

```python
fig, ax = plt.subplots(figsize=(20,7)) 
ax.set_title("Average Price by Monthly Cohorts")
sns.heatmap(data = average_price,
            ax=ax,
            annot=True,
            fmt=".1f",
            vmin= 2,
            vmax = 10,
            cmap = 'Reds',
            yticklabels=retention.index.strftime("%Y-%m"))
```

```python
df['Revenue'] = df['Price'].multiply(df['Quantity'])
grouping = df.groupby(['CohortMonth', 'CohortIndex'])
cohort_data = grouping['Revenue'].sum()
cohort_data = cohort_data.reset_index()

total_spend = cohort_data.pivot(index='CohortMonth', 
                                     columns='CohortIndex', 
                                     values='Revenue')
```

```python
fig, ax = plt.subplots(figsize=(20,7)) 
ax.set_title("Monthly Spend by Cohorts (K$)")
sns.heatmap(data = total_spend.divide(1000),
            ax=ax,
            annot=True,
            fmt=".1f",
            vmin= 10,
            vmax = 100,
            cmap = 'Reds',
            yticklabels=retention.index.strftime("%Y-%m"))
```

## 2. RFM Segmentation

RFM segmentation (R: Recency, F: Frequency, M: Monetary Value) is a common way of behavorial segmentation. Once we calculate these metrics, we can use quantile method to divide the customers into groups based on a complex metrics that we called RFM score.

```python
snapshot_date = dt.datetime(2011,12,10)
# Calculate Recency, Frequency and Monetary value for each customer 
datamart = df.groupby(['Customer ID']).agg({
    'InvoiceDate': {
        'Recency': lambda x: (snapshot_date - x.max()).days,
        'Tenure': lambda x: (snapshot_date - x.min()).days
    },
    'Invoice': 'nunique',
    'Revenue': 'sum'})

# Rename the columns 
datamart.columns = ['Recency','Tenure','Frequency','MonetaryValue']

# Print top 5 rows
datamart.head()
```

```python
# Create labels for Recency and Frequency
r_labels = range(3, 0, -1); f_labels = range(1, 4)

# Assign these labels to three equal percentile groups 
r_groups = pd.qcut(datamart['Recency'], q=3, labels=r_labels)

# Assign these labels to three equal percentile groups 
f_groups = pd.qcut(datamart['Frequency'], q=3, labels=f_labels)

# Create new columns R and F 
datamart = datamart.assign(R=r_groups.values, F=f_groups.values)
```

```python
# Create labels for MonetaryValue
m_labels = range(1, 4)

# Assign these labels to three equal percentile groups 
m_groups = pd.qcut(datamart['MonetaryValue'], q=3, labels=m_labels)

# Create new column M
datamart = datamart.assign(M=m_groups)

# Calculate RFM_Score
datamart['RFM_Score'] = datamart[['R','F','M']].sum(axis=1)

datamart['RFM_Score'].head()
```

```python
datamart['RFM_Score'].min()
```

```python
# Define rfm_level function
def rfm_level(df):
    if df['RFM_Score'] >= 8:
        return 'Top'
    elif ((df['RFM_Score'] >= 6) and (df['RFM_Score'] < 8)):
        return 'Middle'
    else:
        return 'Low'
```

```python
# Create a new variable RFM_Level
datamart['RFM_Level'] = datamart.apply(rfm_level, axis=1)

# Print the header with top 5 rows to the console
datamart.head()
```

```python
# Calculate average values for each RFM_Level, and return a size of each segment 
rfm_level_agg = datamart.groupby('RFM_Level').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
  
  	# Return the size of each segment
    'MonetaryValue': ['mean', 'count']
}).round(1)

# Print the aggregated dataset
rfm_level_agg
```

## 3. Data Pre-processing for Clustering

We use K-means algorithm to build the clustering model because of its effectiveness and effeciency. Before that, we need to deal with the skewness and difference in scale of the input data.

```python
f, axs = plt.subplots(1,4,figsize=(20,5))

plt.subplot(1,4,1); sns.distplot(datamart['Recency'])

plt.subplot(1,4,2); sns.distplot(datamart['Tenure'])

plt.subplot(1,4,3); sns.distplot(datamart['Frequency'])

plt.subplot(1,4,4); sns.distplot(datamart['MonetaryValue'] / 1000) #Unit: thousand 
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
datamart_rfmt = datamart[["Recency","Frequency","MonetaryValue","Tenure"]]

# Unskew the data
datamart_log = np.log(datamart_rfmt).replace([np.inf, -np.inf],np.nan).fillna(0)

# Initialize a standard scaler and fit it
scaler = StandardScaler()
scaler.fit(datamart_log)

# Scale and center the data
datamart_normalized = scaler.transform(datamart_log)

# Create a pandas DataFrame
datamart_normalized = pd.DataFrame(data=datamart_normalized, 
                                   index=datamart_rfmt.index, 
                                   columns=datamart_rfmt.columns)
```

```python
f, axs = plt.subplots(1,4,figsize=(20,5))
# Plot recency distribution
plt.subplot(1, 4, 1); sns.distplot(datamart_normalized['Recency'])

# Plot frequency distribution
plt.subplot(1, 4, 2); sns.distplot(datamart_normalized['Tenure'])

# Plot frequency distribution
plt.subplot(1, 4, 3); sns.distplot(datamart_normalized['Frequency'])

# Plot monetary value distribution
plt.subplot(1, 4, 4); sns.distplot(datamart_normalized['MonetaryValue'])

# Show the plot
plt.show()
```

## 4. Customer Segmentation with K-means

With the pre-processed data, we are able to build a clustering model with K-means algorithm. First, we use visual method of SSE curve the find the best number of clusters.

```python
from sklearn.cluster import KMeans
```

```python
sse = {}
# Fit KMeans and calculate SSE for each k
for k in range(1, 21):
  
    # Initialize KMeans with k clusters
    kmeans = KMeans(n_clusters=k, random_state=1)
    
    # Fit KMeans on the normalized dataset
    kmeans.fit(datamart_normalized)
    
    # Assign sum of squared distances to k element of dictionary
    sse[k] = kmeans.inertia_
```

```python
# Add the plot title "The Elbow Method"
plt.title('The Elbow Method')

# Add X-axis label "k"
plt.xlabel('k')

# Add Y-axis label "SSE"
plt.ylabel('SSE')

# Plot SSE values for each key in the dictionary
sns.pointplot(x=list(sse.keys()), y=list(sse.values()))
plt.show()
```

Now we use the proper number of clusters to build the model.

```python
# Initialize KMeans
kmeans = KMeans(n_clusters=5, random_state=1) 

# Fit k-means clustering on the normalized data set
kmeans.fit(datamart_normalized)

# Extract cluster labels
cluster_labels = kmeans.labels_
```

```python
# Create a DataFrame by adding a new cluster label column
datamart_rfmt_k5 = datamart_rfmt.assign(Cluster=cluster_labels)

# Group the data by cluster
grouped = datamart_rfmt_k5.groupby(['Cluster'])

# Calculate average RFM values and segment sizes per cluster value
grouped.agg({
    'Recency': 'mean',
    'Tenure':'mean',
    'Frequency': 'mean',
    'MonetaryValue': ['mean', 'count']
  }).round(1).sort_values(by=('MonetaryValue','mean'))
```

Now with each segment, we can calculate the key metrics to show the insights.

```python
# Calculate average RFM values for each cluster
cluster_avg = datamart_rfmt_k5.groupby(['Cluster']).mean() 

# Calculate average RFM values for the total customer population
population_avg = datamart_rfmt.mean()

# Calculate relative importance of cluster's attribute value compared to population
relative_imp = cluster_avg / population_avg - 1

# Print relative importance scores rounded to 2 decimals
print(relative_imp.round(2))
```

```python
datamart_normalized['Cluster'] = cluster_labels
datamart_normalized = datamart_normalized[datamart_normalized['Cluster'] != 4]
# Melt the normalized dataset and reset the index
datamart_melt = pd.melt(datamart_normalized.reset_index(), 
                        # Assign CustomerID and Cluster as ID variables
                        id_vars=['Customer ID', 'Cluster'],
                        # Assign RFM values as value variables
                        value_vars=['Recency', 'Tenure','Frequency', 'MonetaryValue'], 
                        # Name the variable and value
                        var_name='Metric', value_name='Value')
```

```python
# Add the plot title
plt.title('Snake plot of normalized variables')

# Add the x axis label
plt.xlabel('Metric')

# Add the y axis label
plt.ylabel('Value')

# Plot a line for each value of the cluster variable
sns.lineplot(data=datamart_melt, x='Metric', y='Value', hue='Cluster')
plt.show()
```

```python
# Initialize a plot with a figure size of 8 by 2 inches 
plt.figure(figsize=(8, 2))

# Add the plot title
plt.title('Relative importance of attributes')

# Plot the heatmap
sns.heatmap(data=relative_imp, annot=True, fmt='.2f', cmap='RdYlGn')
plt.show()
```
